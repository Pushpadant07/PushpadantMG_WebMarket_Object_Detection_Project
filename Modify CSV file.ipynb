# Recreate the csv file into Tensorflow CSV file format

```python
# Importing pandas 
import pandas as pd
```

```python
# reading the CSV file 
x = pd.read_csv("WebMarket_coco_gt_object.csv")
```

```python
# Check first 5 rows of data
x.head()
```

```python
# our data hace 118688 rows and 6 columns
x.shape
```

```python
# extract only `image_name` columns
a=x['image_name']
print(a)
```

```python
#  regular expression hepls to find patterns 
import re
for i in a[0:10]:
    temp = re.findall(r'\d',i) #\d return interger(0-9) and \D return all except integer (0-9)
    print(temp)
```

```python
# apply regex the create new numbers coloumn 
import re
num=[]  
for i in a:
    temp = re.findall(r'\d+', i)
    res = list(map(int, temp))
    num.append(res)
numbers = pd.DataFrame(num)
numbers
```

```python
# Add numbers column into dataframe
x['numbers'] = numbers
x
```

```python
# Add width and height coloumns in dataframe
x["width"]= 2272
x["height"] = 1704
```

```python
x.head()
```

```python
# Sort the whole pandas dataset from one coloumn **numbers** in assending order 
m = x.sort_values("numbers")
print(m)
```

```python
# Drop the number coloumn
m= m.drop(["numbers"],axis=1)
m
```

```python
# Add new column names to dataset
m.columns = ['filename','xmin', 'ymin', 'xmax', 'ymax', 'class', 'width', 'height']
```

```python
m
```

```python
# I have selected all the annotation of 300 images.
db1 = m[0:14269]
db1
```

```python
# Create and save a new CSV file.
from IPython.display import FileLink, FileLinks

db1.to_csv('db1_data.csv', index=False)
```
